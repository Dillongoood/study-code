```python

import os
import re
from datetime import datetime, timedelta

# 指定文件夹路径
folder_path = 'D:/RadarData/2022.678nc/HSR'

# 定义文件名的正则表达式模式
pattern = re.compile(r'Z9010_(\d{8})(\d{6})Z_HSR_00_033')

# 定义起始时间和时间间隔
# 定义起始时间和时间间隔
start_time = datetime(2022, 6, 1, 0, 0, 0)
end_time = datetime(2022, 8, 31, 23, 59, 59)  # 调整为包括指定文件的结束时间
time_interval = timedelta(minutes=6)

# 遍历文件夹中的文件
for filename in os.listdir(folder_path):
    print(f"Processing file: {filename}")
    match = pattern.match(filename)
    if match:
        date_str, time_str = match.groups()
        file_time = datetime.strptime(date_str + time_str, '%Y%m%d%H%M%S').replace(second=0)

        # 检查时间范围
        if start_time <= file_time <= end_time:
            # 生成新的文件名
            new_filename = f"Z9010_{file_time.strftime('%Y%m%d%H%M00')}Z_HSR_00_033.nc"
            
            # 重命名文件
            old_path = os.path.join(folder_path, filename)
            new_path = os.path.join(folder_path, new_filename)
            os.rename(old_path, new_path)
            
            print(f"Renamed file: {filename} to {new_filename}, Time: {file_time}")
        else:
            print(f"Skipping file: {filename}, Time: {file_time}")
    else:
        print(f"File {filename} does not match the pattern.")



```
